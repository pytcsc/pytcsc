```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Salary_Data.csv")
```

```python
degree = 2
coeffs = np.polyfit(data['YearsExperience'], data['Salary'], degree)  # Correct reference to column names
poly_func = np.poly1d(coeffs)

# Generate points for the fit line within the range of YearsExperience
x_fit = np.linspace(data['YearsExperience'].min(), data['YearsExperience'].max(), 100)
y_fit = poly_func(x_fit)

# Plotting the data points and the polynomial fit
plt.scatter(data['YearsExperience'], data['Salary'], label="Data Points")
plt.plot(x_fit, y_fit, label=f'Polynomial Fit (Degree {degree})', color='r')
plt.xlabel('Years of Experience')
plt.ylabel("Salary")
plt.title("Polynomial Regression of Salary vs. Experience")
plt.legend()
plt.show()
```
